<h1>Section A: Data Exploration</h1>

```python
import numpy as np 

import matplotlib.pyplot as plt 

import pandas as pd

df = pd.read_csv("students_large.csv")
df
```

```python
# Q1 Load dataset and inspect:
#Shape
#Columns
#Data types  

df.shape
df.columns.tolist()
df.dtypes


```

```python
# Q2 Check:

#Missing values
#Duplicate records 

df.isnull().sum() 
df.duplicated()
```

<h1>Section B: Feature Engineering</h1>

```python
# Q3 Create:
#Total_Marks 

df["Total_Marks"] = df["Math"] + df["Science"] + df["English"] 
print(df)
```

```python
# Q4 Create:
#Percentage 

df["Percentage"] = (df["Total_Marks"]/300) * 100 
print(df) 
```

```python
# Q5 Create Grade column:

# Percentage	Grade
# >=90	         A
# >=80	         B
# >=70	         C
# >=60	         D
# <60	         F 

import numpy as np

df["Grade"] = np.where(df["Percentage"] >= 90, "A",
                np.where(df["Percentage"] >= 80, "B",
                np.where(df["Percentage"] >= 70, "C",
                np.where(df["Percentage"] >= 60, "D", "F"))))
                       
print(df)
                       
```

<h1>Section C: Statistical Analysis</h1>

```python
# Q6 Calculate:

#Average Percentage
#Highest Percentage
#Lowest Percentage
#Standard Deviation 

df["Percentage"].mean()
df["Percentage"].max()
df["Percentage"].min() 
df["Percentage"].std() 
```

```python
# Q7 Find:

#Top 10 Students
#Bottom 10 Students 

#Top 10 Students
top_10 = df.sort_values(by="Percentage", ascending=False).head(10) 
print(top_10)

#Bottom 10 Students 
top_10 = df.sort_values(by="Percentage", ascending=True).head(10) 
print(top_10) 

```

```python
# Q8 Subject-wise average:

#Math
#Science
#English 

df[["Math", "Science", "English"]].mean() 
```

<h1>Section D: Correlation Analysis</h1>

```python
# Q9 Find correlation between:
#Attendance and Percentage

#Q10 Does higher attendance generally lead to better scores?
#Support with data. 

atten_per = df["Attendance"].corr(df["Percentage"])
print(atten_per) 

#-0.0000 ka matlab basically 0 ke barabar correlation hai.

#📌 Meaning:
#Attendance aur Percentage ke beech koi meaningful relationship nahi hai
#Na positive effect hai, na negative effect
```

```python
# Q10 Does higher attendance generally lead to better scores?
# Support with data. 

✔️ Final Answer (write in exam):

The correlation between Attendance and Percentage is approximately 0, which indicates that there is no significant relationship between attendance and scores in this dataset.
Therefore, higher attendance does not necessarily lead to better scores
```

<h1>Section E: Visualization</h1>

```python
# Chart 1 Histogram
#Percentage Distribution 

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df["Percentage"], bins=10)

plt.title("Percentage Distribution")
plt.xlabel("Percentage")
plt.ylabel("Number of Students")

plt.show() 
```

```python
#Chart 2
#Scatter Plot
#Attendance vs Percentage 

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(df["Attendance"] , df["Percentage"])

plt.title("Percentage Distribution")
plt.xlabel("Attendance")
plt.ylabel("Percentage")

plt.show() 
```

```python
# Chart 3
#Bar Chart
#Average Marks by Subject 

import matplotlib.pyplot as plt

avg_subject = df[["Math", "Science", "English"]].mean()
plt.figure(figsize=(8,5))
plt.bar(avg_subject.index, avg_subject.values)

plt.title("Percentage Distribution")
plt.xlabel("index")
plt.ylabel("")

plt.show() 
```
